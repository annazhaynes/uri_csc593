I wanted to be able to vizualize the number of crimes grouped by different columns in my dataset. This will help me to determine the frequency of certain crimes, where they are committed and when. I could go further and combine these, which day had the most assault cases? That sort of thing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook

bostoncrime = pd.read_excel('../dataset/cleancrimeboston.xlsx', sheet_name=0)
```

```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

fig, ax = plt.subplots(figsize=(18,15))

#creates a frequency bar graph based on the offence code group
crimeplot = bostoncrime['OFFENSE_CODE_GROUP'].value_counts().plot(ax=ax, kind='bar')

#sets the labels of axis and title
plt.xlabel('Type of Crime')
plt.ylabel('Times Committed')
plt.title('Frequency of Committed Crimes by Category')

#formatting the tick marks on the y-axis as well as changing the step size
start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(start, end, 500))
ax.yaxis.set_major_locator(MultipleLocator(500))
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))
ax.yaxis.set_minor_locator(MultipleLocator(250))

#adding a grid to read the graph easier
plt.grid(axis = 'y', linestyle='--')
```

```python
fig, ax = plt.subplots(figsize=(18,15))
bostoncrime['DISTRICT'].value_counts().plot(ax=ax, kind='bar')
plt.xlabel('District')
plt.ylabel('Crime Committed')
plt.title('Frequency of Committed Crimes by District')
start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(start, end, 500))
ax.yaxis.set_major_locator(MultipleLocator(500))
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))
ax.yaxis.set_minor_locator(MultipleLocator(250))
plt.grid(axis = 'y', linestyle='--')
```

```python
fig, ax = plt.subplots(figsize=(18,15))
bostoncrime['DAY_OF_WEEK'].value_counts().plot(ax=ax, kind='bar')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes Committed')
plt.title('Frequency of Committed Crimes by Day of the Week')
start, end = ax.get_ylim()
ax.yaxis.set_ticks(np.arange(start, end, 1000))
ax.yaxis.set_major_locator(MultipleLocator(1000))
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))
ax.yaxis.set_minor_locator(MultipleLocator(250))
plt.grid(axis = 'y', linestyle='--')
```
